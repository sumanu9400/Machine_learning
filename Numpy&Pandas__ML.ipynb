```python
import numpy as np
numbers=[99,86,87,88,111,86,103,87,94,78,77,85,86]
x=np.mean(numbers)
print(x)
```

```python
import numpy as np
speed=[99,86,87,88,111,86,103,87,94,78,77,85,86]
x=np.median(speed)
print(x)
```

```python
from scify import stats
speed=[99,86,87,88,111,86,103,87,94,78,77,85,86]
x=stats.mode(speed)
print(x)
```
